```python
import pandas as pd
import re
product_df=pd.read_csv("Master products list(USD) - Clawset Products list.csv", encoding='latin')
product_df.head()
```

```python
product_df=product_df[['SKU', 'Categories','Products name (Eng)']]
product_df.rename(columns = {'Products name (Eng)':'Title'}, inplace = True)

product_df
```

```python
product_df['Title'] = product_df['Title'].fillna('') + ' ' + product_df['Categories'].fillna('')
```

```python
# keywords_to_remove =['Green','Pink','Rainbow','Black','Purple','Yellow','Brown','Red','Navy Blue','Cream','White','Grey', 'Orange', 'Light Blue']
# for keyword in keywords_to_remove:
#     product_df['Title'] = product_df['Title'].str.replace(keyword, ' ')

def remove_keywords(input_string):
    keywords_to_remove = ['Green', 'Pink', 'Rainbow', 'Black', 'Purple', 'Yellow', 'Brown', 'Red', 'Navy Blue', 'Cream', 'White', 'Grey', 'Orange', 'Light Blue', 'Blue']
    
    for keyword in keywords_to_remove:
        input_string = input_string.replace(keyword, "")
    return input_string
product_df['Title'] = product_df['Title'].apply(remove_keywords)
```

```python
def remove_multiple_spaces(text):
    # Remove multiple spaces and strip leading/trailing spaces
    text = text.strip()
    text = re.sub(r'\s+', ' ', text)
    return text
product_df['Title'] = product_df['Title'].apply(remove_multiple_spaces)
```
