### Streamlit App for Recommendation System

```python
%%writefile app.py
import streamlit as st
import spacy
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

#nlp = spacy.load("clawset_spacy_model")
nlp = spacy.load("en_core_web_md")
df = pd.read_pickle("clawset_product_descriptions.pkl")

# Preprocess and vectorize product descriptions
processed_descriptions = [nlp(desc.lower()) for desc in df['Description']]

# Compute similarity matrix
similarity_matrix = cosine_similarity([doc.vector for doc in processed_descriptions])

# Generate recommendations
def get_recommendations(query, k=5):
    if query:
      query_doc = nlp(query.lower())
      query_vector = query_doc.vector
      sim_scores = cosine_similarity([query_vector], [doc.vector for doc in processed_descriptions])[0]
      top_indices = sim_scores.argsort()[-k:][::-1]  # Get top k similar descriptions

      recommendations = [(df.iloc[i]['Description'], sim_scores[i]) for i in top_indices]
      return recommendations
    else:
      return []


st.title('Product Recommendation System')
query=st.text_input('Search',placeholder='Enter Products')

# recommended_products = get_recommendations(query)
#for product in recommended_products:
#    st.write(product)

if query:
    recommended_products = get_recommendations(query)
    if recommended_products:
        for product in recommended_products:
            st.write(product)
    else:
        st.write("No recommendations found.")
```

```python
# df = df[['SKU', 'Products name (Eng)', 'Categories']]
# df.head()
# df['']

def clean_text(text):
    # text = "Birthday baby - Party hat- Pink, Size M"
    text = re.sub(r', Size [ML]', '', text)
    
    # Remove special characters like \n
    text = re.sub(r'[\n]', '', text)
    
    # Split the cleaned text into separate lines again for readability
    text = re.sub(r'(\d{3})', r'\n\1', text).strip()
    text = re.sub(r'["\,]', ' ', text)
    return text

df['Products name (Eng)'] = df['Products name (Eng)'].apply(clean_text)
df['Products name (Eng)']
```
